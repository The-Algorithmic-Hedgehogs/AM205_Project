## This notebook to simulate the first part of our model with 100hrs at 0.1 time step.

Collecting P, C, R, S, F2, F3, A every half-hour. Collecting S every 0.5min.

```python
#import necessary packages
import numpy as np
from scipy.optimize import minimize
import matplotlib
import numpy as np
import matplotlib.cm as cm
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
import math
import scipy.optimize as optimize
from scipy import integrate
```

```python
#Define parameters from the Elowitz paper
alpha_ptc = 0.024
gamma_p = 0.006
gamma_c = 0.1067
K_h = 1.8124
k_off = 0.9157
k_on = K_h * k_off
K_p = 0.01
n_p = 1.3007
alpha_f = 0.7610
gamma_f = 0.002
alpha_a = 1.9996
alpha_R2 = 0.00326
alpha_R3 = 0.0315
gamma_a = 0.02
gamma_R = 0.002
K_gli = 0.1501
alpha_GBS = 0.0022
alpha_GBS0 = 1.384e-4
D_diff = 52.3942
alpha_HH = 0.4898
gamma_H = 7.644e-6
beta = 16
m = 2.5595
beta_fl = 2.3797
t_dr = 30
```

```python
#function to create the differentiation matrix needed for Hedgehog diffusion.
def create_d_matrix(n):
    h = 1/(n-1)
    D = -2*np.identity(n) 
    for i in range(n-1):
        D[i+1, i] = 1
        D[i, i+1] = 1       
    D[n-1] = 0
    D[0,1] = 2    
    return D
```

```python
#function to calculate the adaptive behavior of the pathway. Used in the functions update_F2 and update_F3 below.
def AF(t):
    if t < t_dr:
        return alpha_f
    elif t >= t_dr and t < 2.5*t_dr:
        return alpha_f*(2.5*t_dr - t)/(1.5*t_dr)
    else:
        return 0

#function to calculate H + dH/dt. This function returns the H values at each point at time n+1
def update_H(H, P, C, t, D):
    n = len(H)
    h = 1/(n-1)
    alpha = (D_diff*t)/(h**2)
    M = H + (k_off*C)
    M[0] += (alpha_HH/D_diff)*2*h*alpha
    
    Differentiation_Matrix = D
    matrix_HP = np.matmul(k_off*P, np.identity(n))
    matrix_gamma = gamma_H * np.identity(n)
    
    A = np.identity(n) - ((alpha*Differentiation_Matrix) - matrix_HP - matrix_gamma)
    future_H = np.linalg.solve(A, M)
    
    future_H[future_H < 0] = 0
    return future_H

#function to calculate phiAR
def phi_AR(A,R):
    num = A**2 + K_gli*A
    den = A**2 + 2*K_gli*A + 2*K_gli*R + 2*A*R + R**2 + K_gli**2
    return num/den

#function to calculate P + dP/dt. This function returns the P values at each point at time n+1
def update_P(H, A, P, R, C, t):
    phi = phi_AR(A,R)
    n = len(H)
    delta = ((alpha_ptc)*np.ones(n)) + ((alpha_ptc)*beta*phi) - (gamma_p*P) - (k_on*H*P) + (k_off*C)
    future_P = P + delta*t
    future_P[future_P < 0] = 0
    return future_P

#function to calculate C + dC/dt. This function returns the C values at each point at time n+1
def update_C(H, P, C, t):
    n = len(H)
    delta = (k_on*H*P) - (k_off*C) - (gamma_c*C)
    future_C = C + delta*t
    future_C[future_C < 0] = 0      
    return future_C

#function to calculate F2 + dF2/dt. This function returns the F2 values at each point at time n+1
def update_F2(A,R,P,F2,t, elapsed_time):
    af = AF(elapsed_time)
    phi = phi_AR(A,R)
    g1 = af*(1+beta_fl*phi)
    g2 = alpha_a*((K_p**n_p)/((K_p**n_p)+(P**n_p))) + alpha_R2 + gamma_f
    delta = (g1 - g2*F2)
    future_F2 = F2 + delta*t
    future_F2[future_F2 < 0] = 0
    return future_F2

#function to calculate F3 + dF3/dt. This function returns the F3 values at each point at time n+1
def update_F3(P,F3,t, elapsed_time):
    af = AF(elapsed_time)
    delta = af - (alpha_a*((K_p**n_p)/((K_p**n_p)+(P**n_p))) + alpha_R3 + gamma_f)*F3
    future_F3 = F3 + delta*t
    future_F3[future_F3 < 0] = 0
    return future_F3   

#function to calculate R + dR/dt. This function returns the R values at each point at time n+1
def update_R(F2, F3, R, t):
    delta = alpha_R2*F2 + alpha_R3*F3 - gamma_R*R
    future_R = R + delta*t
    future_R[future_R < 0] = 0
    return future_R

#function to calculate A + dA/dt. This function returns the A values at each point at time n+1
def update_A(P, F2, F3, A, t):
    delta = alpha_a*((K_p**n_p)/((K_p**n_p)+(P**n_p)))*(F2+F3) - gamma_a*A
    future_A = A + delta*t
    future_A[future_A < 0] = 0
    return future_A

#function to calculate S + dS/dt. This function returns the S values at each point at time n+1
def update_S(R, A, S, t):
    gamma_S = 0
    phi = phi_AR(A,R)
    n = len(R)
    delta = (alpha_GBS0*np.ones(n)) + (alpha_GBS*phi) - (gamma_S * S)
    future_S = S + delta*t
    future_S[future_S < 0] = 0
    return future_S, delta
```

```python
#function to perform simulation given a final time in number of minutes, time_step in number of minutes, and n grid points.
def integrate_solution(final_time, time_step, n):
    H = np.zeros(n)
    C = np.zeros(n)
    S = np.zeros(n)
    F2 = 125*np.ones(n)
    F3 = 22*np.ones(n)
    A = 6*np.ones(n)
    P = 4*np.ones(n)
    R = 550*np.ones(n)
    time = 0; #start time
    D = create_d_matrix(n); #create differentiation matrix for diffusion
    min_to_collect_at = 30;
    hrs = int( np.floor( final_time / min_to_collect_at)); #number of hours we will simulate over
    
    #make arrays to hold the values of each protein at each grid point at each hour
    H_vals = np.zeros((hrs+1,n));
    H_vals[0] = H; #initialize at t=0
    C_vals = np.zeros((hrs+1,n));
    C_vals[0] = C; #initialize at t=0
    S_vals = []; #need this to be a list since we are appending at every half-min
    S_vals.append(S); #initialize at t=0
    P_vals = np.zeros((hrs+1,n));
    P_vals[0] = P; #initialize at t=0
    R_vals = np.zeros((hrs+1,n));
    R_vals[0] = R; #initialize at t=0
    F2_vals = np.zeros((hrs+1,n));
    F2_vals[0] = F2; #initialize at t=0
    F3_vals = np.zeros((hrs+1,n));
    F3_vals[0] = F3; #initialize at t=0
    A_vals = np.zeros((hrs+1,n));
    A_vals[0] = A; #initialize at t=0
    delta_S_vals = []; #need this to be a list since we are appending at every half-min
    S_vals.append(S); #initialize at t=0
    
    #need plus one because there is some rounding error in the addition for the time step
    steps_each_hr = min_to_collect_at / time_step + 1; 
    print("Number of steps to take each hour: ", steps_each_hr);
    steps_each_half_min = 0.5 / time_step; #should be taking sample of S 
    print("Number of steps to take to reach 0.5min : ", steps_each_half_min);
    steps = 0;
    
    #keep looping and updating protein concentrations while time is below threshold
    while time<final_time:
        #calculate the updates for each protein
        P_upd = update_P(H, A, P, R, C, time_step);
        C_upd = update_C(H, P, C, time_step);
        F2_upd = update_F2(A,R,P,F2,time_step, time/60);
        F3_upd = update_F3(P,F3,time_step, time/60);
        R_upd = update_R(F2, F3, R, time_step);
        A_upd = update_A(P, F2, F3, A, time_step);
        S_upd, delta_S = update_S(R, A, S, time_step);
        
        H_upd = update_H(H, P, C, time_step, D);
        
        #reassign each protein to new concentrations
        H = H_upd.copy();
        P = P_upd.copy();
        C = C_upd.copy();
        F2 = F2_upd.copy();
        F3 = F3_upd.copy();
        R = R_upd.copy();
        A = A_upd.copy();
        S = S_upd.copy();
        
        time += time_step #increment total time
        steps +=1;
        if steps % steps_each_half_min == 0: #save S if we've reached 0.5 min
            S_vals.append(S);
            delta_S_vals.append(delta_S)
        if steps == steps_each_hr : #we've reached an hour
            #store protein concentrations at each grid point for this hour
            index = int(time / min_to_collect_at);
            print("Index to insert at:", index, " at time ", time);
            H_vals[index] = H;
            P_vals[index] = P;
            C_vals[index] = C;
            F2_vals[index] = F2;
            F3_vals[index] = F3;
            R_vals[index] = R;
            A_vals[index] = A;
            steps = 0; #reinitialize steps
            
    #add results from last time step to matrices
    H_vals[-1] = H;
    P_vals[-1] = P;
    C_vals[-1] = C;
    F2_vals[-1] = F2;
    F3_vals[-1] = F3;
    R_vals[-1] = R;
    A_vals[-1] = A;
    S_vals.append(S);
        
    #return matrices for each protein    
    return H_vals, P_vals, C_vals, R_vals, S_vals, F2_vals, F3_vals, A_vals, delta_S_vals;
```

```python
from time import clock

c = clock();
hrs_to_sim = 100;
time_step = 0.1;
H, P, C, R, S, F2, F3, A, delta_S = integrate_solution(hrs_to_sim*60, time_step, 420);
stop = clock();

print("Time: {0} min".format( (stop-c)/60 ) );
```

```python
#function to normalize S values by substracting the S values at each time point by minimum S for that time point
def normalize_S(S):
    normed_S = S.copy()
    normed_S = normed_S - np.min(normed_S)
    return normed_S
```
