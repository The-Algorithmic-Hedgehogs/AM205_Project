```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
shh_vals = np.loadtxt("./100_hr_sim/H_100hr_sim.csv", delimiter=',');
print(shh_vals.shape)
```

```python
fig = plt.figure(figsize=(12,7));
ax = plt.subplot(1,1,1);
hrs = [1,2,3,4, 35, 37];
for h in hrs:
    ax.plot(shh_vals[h], label="SHH, hour={0}".format(h));
    
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5), fontsize= 12); 

plt.show();
```
