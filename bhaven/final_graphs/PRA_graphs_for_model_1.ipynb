```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
#get P data
p_data = np.loadtxt("../../Hugo/P_100hr_sim_final.csv", delimiter=',');

print(p_data.shape); #make sure shape is correct
```

```python
#Patched levels across space
fig = plt.figure(figsize=(10,7));
ax = plt.subplot(1,1,1);
fontsize = 16;
ax.tick_params(labelsize=fontsize);
ax.set_xlabel('x-position ($\mu{m}$)', fontsize=fontsize);
ax.set_ylabel('Patched levels', fontsize=fontsize);
# ax.set_title("Response across space", fontsize=fontsize+2)


hrs = [1, 5, 10, 20, 40, 60, 80, 100] #hours to plot 
for h in hrs:
    index = 2 * h;
    ax.plot(p_data[index], label="hour={0}".format(h));
    
ax.legend(loc='best', fontsize= 12); 

plt.show();
```

```python
#Patched across time
fig = plt.figure(figsize=(12,7));
ax = plt.subplot(1,1,1);
fontsize = 16;
ax.tick_params(labelsize=fontsize);
ax.set_xlabel('time (hr)', fontsize=fontsize);
ax.set_ylabel('Patched levels', fontsize=fontsize);
# ax.set_title("Response across time", fontsize=fontsize+2);

ax.set_xticks(np.arange(0, 201, step=40))
ax.set_xticklabels(labels=[0,20,40,60,80,100])

xp = [1, 50, 100, 150, 200, 250] #x-points to plot 

for x in xp:
    ax.plot(p_data[:,x], label="x={0} $\mu m$".format(x));
    
ax.legend(loc='upper right', fontsize= 12); 

plt.show();
```
