```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
#get H data
h_data = np.loadtxt("../../Hugo/H_100hr_sim_final.csv", delimiter=',');

print(h_data.shape); #make sure shape is correct
```

```python
#Hedgehog levels across space
fig = plt.figure(figsize=(10,7));
ax = plt.subplot(1,1,1);
fontsize = 16;
ax.tick_params(labelsize=fontsize);
ax.set_xlabel('x-position ($\mu{m}$)', fontsize=fontsize);
ax.set_ylabel('Hedgehog levels', fontsize=fontsize);

hrs = [1, 5, 10, 20, 40, 60, 80, 100] #hours to plot 
for h in hrs:
    index = 2 * h;
    ax.plot(h_data[index], label="hour={0}".format(h));
    
ax.legend(loc='best', fontsize= 12); 

plt.show();
```

```python
#Hedgehog across time
fig = plt.figure(figsize=(12,7));
ax = plt.subplot(1,1,1);
fontsize = 16;
ax.tick_params(labelsize=fontsize);
ax.set_xlabel('time (hr)', fontsize=fontsize);
ax.set_ylabel('Hedgehog levels', fontsize=fontsize);

ax.set_xticks(np.arange(0, 201, step=40))
ax.set_xticklabels(labels=[0,20,40,60,80,100])

xp = [1, 50, 100, 150, 200, 250] #x-points to plot 

for x in xp:
    ax.plot(h_data[:,x], label="x={0} $\mu m$".format(x));
    
ax.legend(loc='upper right', fontsize= 12); 

plt.show();
```
